# NoShow - Feature Selection Strategy

```python
import numpy as np
import pandas as pd
import seaborn as sns
import pyMechkar as pyMechkar
```

```python
df = pd.read_csv("c:/Users/Mayer/Documents/DataScience/project/Noshows/NoShowClusters.csv")
df.head()
```

```python
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
df['Client_Code'] =  df['Client_Code'].astype('category')
df['Booking_Date'] = pd.to_datetime(df['Booking_Date'])
df[df.columns[2:8]] = df[df.columns[2:8]].astype('category')
df['cancel_days'] =  df['cancel_days'].astype('category')
df[['quantity','avg_receipt','avg_quantity']] =  df[['quantity','avg_receipt','avg_quantity']].astype('category')
df['NoShow'] = df['NoShow'].astype('category')
df[df.columns[10:55]] = df[df.columns[10:55]].astype('category')
df[df.columns[60:65]] = df[df.columns[60:65]].astype('category')


df.dtypes[df.columns[4:65]]
```

converting booking_Date to 'int'

```python
df['Booking_Date']=df.Booking_Date.apply(lambda x: x.strftime('%Y%m%d')).astype('int64')
df.dtypes[df.columns[1:2]]
```

```python
mid = df['NoShow']
df.drop(labels=['NoShow'], axis=1, inplace = True)
df.insert(63, 'NoShow', mid)
df
```

```python
varSel = pd.DataFrame({'Variable':df.columns[0:63]})
varSel
```

## Univariable Analysis

```python
from importlib import reload
from pyMechkar import analysis
```

```python
tab1 = analysis.Table1(data=df, y='NoShow')
tab1
```

```python
tab1[tab1['p_value']<0.05]
```

```python
vn1 = tab1.loc[tab1['p_value']<0.05,'Variables'].unique()
print(len(vn1))
vn1
```

We will add these variables to our variable selection table

```python
varSel['Univarable'] = 0
varSel.loc[varSel['Variable'].isin(vn1), 'Univarable'] = 1
varSel
```

## Multivariable Analysis

```python
## remove unnecessary vars
X = df.loc[:,df.columns[1:62]]
X.head()
```

```python
y = df.loc[:,df.columns[63]]
print([X.shape,y.shape])
```

### Variable Selection using LASSO (L1 penalization)

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lassomod = Lasso(alpha=0.01).fit(X, y)
```

```python
model = SelectFromModel(lassomod, prefit=True)
model.get_support()
```

```python
varSel['Lasso'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

rfmod = RandomForestClassifier().fit(X, y)
#rfmod.feature_importances_ 
```

```python
model = SelectFromModel(rfmod, prefit=True)
model.get_support()
```

```python
varSel['RandomForest'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Gradient Boosting classification

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.feature_selection import SelectFromModel

gbmod = GradientBoostingClassifier().fit(X, y)
```

```python
model = SelectFromModel(gbmod, prefit=True)
model.get_support()
```

```python
varSel['GradientBoost'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using SVM classification

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel

svmmod = LinearSVC(C=0.01, penalty="l1",dual=False).fit(X, y)
```

```python
model = SelectFromModel(svmmod, prefit=True)
model.get_support()
```

```python
varSel['SVM'] = model.get_support().astype('int64')
varSel
```

### Summarization and Selection of Variables 

```python
varSel['Sum'] =  np.sum(varSel,axis=1)
varSel
```

```python
varSel.groupby('Sum')['Variable'].count()
```

We can now decide a threshold for selecting our variables!
