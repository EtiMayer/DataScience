## Multivariable Analysis

```python
#class_weight = 'balanced'
#class_weight = {0: 0.5, 1: 5}
class_weight = {0: 1, 1: 3}
```

### Variable Selection using LASSO (L1 penalization)

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lassomod = Lasso(alpha=0.01,random_state=1065).fit(X,y)
```

```python
model = SelectFromModel(lassomod, prefit=True)
model.get_support()
```

```python
varSel['Lasso'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Ridge (L2 penalization)

```python
from sklearn.linear_model import Ridge
from sklearn.feature_selection import SelectFromModel

ridgemod = Ridge(alpha=0.01,random_state=1065).fit(X,y)
```

```python
model = SelectFromModel(ridgemod, prefit=True)
model.get_support()
```

```python
varSel['Ridge'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Naive Bayes

```python
# CNB is particularly suited for imbalanced data sets.
from sklearn.naive_bayes import ComplementNB

nbmod = ComplementNB(norm=True).fit(X,y)
```

```python
model = SelectFromModel(nbmod, prefit=True)
model.get_support()
```

```python
varSel['ComplementNB'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfmod = RandomForestClassifier(class_weight = class_weight,random_state=1065).fit(X, y)
#rfmod = RandomForestClassifier(class_weight = 'balanced').fit(X, y)
```

```python
model = SelectFromModel(rfmod, prefit=True)
model.get_support()
```

```python
varSel['RandomForest'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Classification Trees (CART)

```python
from sklearn.tree import DecisionTreeClassifier


cartmod = DecisionTreeClassifier(class_weight=class_weight,random_state=1065).fit(X,y)
#cartmod = DecisionTreeClassifier(class_weight='balanced').fit(X,y)
```

```python
model = SelectFromModel(cartmod, prefit=True)
model.get_support()
```

```python
varSel['CART'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using SVM classification

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel

svmmod = LinearSVC(C=0.01, penalty="l1",dual=False, class_weight=class_weight,random_state=1065).fit(X, y)
#svmmod = LinearSVC(C=0.01, penalty="l1",dual=False, class_weight='balanced').fit(X, y)
```

```python
model = SelectFromModel(svmmod, prefit=True)
model.get_support()
```

```python
varSel['SVM'] = model.get_support().astype('int64')
varSel
```
