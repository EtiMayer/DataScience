# NoShow - Feature Selection Strategy

```python
import numpy as np
import pandas as pd
import seaborn as sns
import pyMechkar as pyMechkar
import sklearn
from numpy.random import seed
```

```python
#random.seed(1)
#random.seed(844)
##random.seed(1065)
#random.seed(632)
seed(952)
#seed(500)
```

```python
df = pd.read_csv("c:/Users/Mayer/Documents/DataScience/project/Noshows/NoShowClusters.csv",index_col=False)
df.head()
```

```python
df[df.columns[2:8]] = df[df.columns[2:8]].astype('int64')
df['cancel_days'] =  df['cancel_days'].astype('category')
df[['quantity','avg_receipt','avg_quantity']] =  df[['quantity','avg_receipt','avg_quantity']].astype('category')
df['NoShow'] = df['NoShow'].astype('int64')
df[df.columns[10:55]] = df[df.columns[10:55]].astype('int64')
df[df.columns[60:65]] = df[df.columns[60:65]].astype('int64')

df.dtypes[df.columns[1:65]]
```

```python
mid = df['NoShow']
df.drop(labels=['NoShow'], axis=1, inplace = True)
df.insert(2, 'NoShow', mid)
df
```

#### One-hot encoding for categorical features

```python
#pip install category_encoders
from category_encoders import OneHotEncoder as enc
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
```

```python
#enc = OneHotEncoder(handle_unknown='ignore')
```

```python
enc = enc(verbose=0, cols=None, drop_invariant=False, return_df=True, handle_missing='value', handle_unknown='value', use_cat_names=False)
```

```python
## remove unnecessary vars
dfenc = df.loc[:,df.columns[2:65]]
```

```python
dfenc.head(3)
```

```python
dfenc = enc.fit_transform(dfenc)
dfenc.head(3)
```

```python
# merging Client_Code to dfenc
#Client_Code = df.loc[:,df.columns[0]]
#Client_Code = Client_Code.to_frame()
#dfenc = pd.concat([Client_Code, dfenc], axis=1, sort=False)
#dfenc.head()
```

```python
## remove unnecessary vars
X = dfenc.loc[:,dfenc.columns[1:82]]
X.head()
```

```python
y = dfenc.loc[:,dfenc.columns[0]]
print([X.shape,y.shape])
```

```python
VarName=enc.get_feature_names()
VarName.remove('NoShow')
VarName
```

```python
# VarSel dataframe
varSel = pd.DataFrame({'Variable':VarName})
varSel
```

## Univariable Analysis

```python
from importlib import reload
from pyMechkar import analysis
```

```python
tab1 = analysis.Table1(data=dfenc,y='NoShow')
tab1
```

```python
tab1[tab1['p_value']<0.05]
```

```python
vn1 = tab1.loc[tab1['p_value']<0.05,'Variables'].unique()
print(len(vn1))
vn1
```

We will add these variables to our variable selection table

```python
varSel['Univariable'] = 0
varSel.loc[varSel['Variable'].isin(vn1), 'Univariable'] = 1
varSel
```

## Multivariable Analysis

```python
#class_weight = 'balanced'
class_weight = {0: 0.5, 1: 5}
#class_weight = {0: 1, 1: 3}
```

### Variable Selection using LASSO (L1 penalization)

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lassomod = Lasso(alpha=0.01).fit(X,y)
```

```python
model = SelectFromModel(lassomod, prefit=True)
model.get_support()
```

```python
varSel['Lasso'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Ridge (L2 penalization)

```python
from sklearn.linear_model import Ridge
from sklearn.feature_selection import SelectFromModel

ridgemod = Ridge(alpha=0.01).fit(X,y)
```

```python
model = SelectFromModel(ridgemod, prefit=True)
model.get_support()
```

```python
varSel['Ridge'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Naive Bayes

```python
# CNB is particularly suited for imbalanced data sets.
from sklearn.naive_bayes import ComplementNB

nbmod = ComplementNB().fit(X,y)
```

```python
model = SelectFromModel(nbmod, prefit=True)
model.get_support()
```

```python
varSel['ComplementNB'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfmod = RandomForestClassifier(class_weight = class_weight).fit(X, y)
#rfmod = RandomForestClassifier(class_weight = 'balanced').fit(X, y)
```

```python
model = SelectFromModel(rfmod, prefit=True)
model.get_support()
```

```python
varSel['RandomForest'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Classification Trees (CART)

```python
from sklearn.tree import DecisionTreeClassifier


cartmod = DecisionTreeClassifier(class_weight=class_weight).fit(X,y)
#cartmod = DecisionTreeClassifier(class_weight='balanced').fit(X,y)
```

```python
model = SelectFromModel(cartmod, prefit=True)
model.get_support()
```

```python
varSel['CART'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using Gradient Boosting classification

```python
from sklearn.ensemble import GradientBoostingClassifier
#sample_weights = np.zeros(len(y))
#sample_weights[y == 0] = 1
#sample_weights[y == 1] = 3

gbmod = GradientBoostingClassifier().fit(X, y)
```

```python
model = SelectFromModel(gbmod, prefit=True)
model.get_support()
```

```python
varSel['GradientBoost'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using ADAboost classification

```python
from sklearn.ensemble import AdaBoostClassifier

adabmod = AdaBoostClassifier().fit(X, y)
```

```python
model = SelectFromModel(adabmod, prefit=True)
model.get_support()
```

```python
varSel['ADAboost'] = model.get_support().astype('int64')
varSel
```

### Variable Selection using SVM classification

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel

svmmod = LinearSVC(C=0.01, penalty="l1",dual=False, class_weight=class_weight).fit(X, y)
#svmmod = LinearSVC(C=0.01, penalty="l1",dual=False, class_weight='balanced').fit(X, y)
```

```python
model = SelectFromModel(svmmod, prefit=True)
model.get_support()
```

```python
varSel['SVM'] = model.get_support().astype('int64')
varSel
```
