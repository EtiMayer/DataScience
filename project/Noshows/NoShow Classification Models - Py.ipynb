## CV strategy Selection

Some classification problems can exhibit a large imbalance in the distribution of the target classes: for instance there could be several times more negative samples than positive samples. In such cases it is recommended to use stratified sampling as implemented in StratifiedKFold and StratifiedShuffleSplit to ensure that relative class frequencies is approximately preserved in each train and validation fold.

```python
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
skf = StratifiedKFold(n_splits=3, shuffle=False, random_state=952)

skf.split(X_train, y_train)
```

```python
sss = StratifiedShuffleSplit(n_splits=3, test_size=0.5, random_state=952)
sss.split(X_train, y_train)
```

```python
from  sklearn.linear_model import LogisticRegressionCV
```

```python
mod1 = LogisticRegressionCV(cv=skf,solver='liblinear',random_state=952).fit(X_train,y_train)
mod2 = LogisticRegressionCV(cv=sss,solver='liblinear',random_state=952).fit(X_train,y_train)
```

```python
scores_skf = cross_validate(mod1, X_train, y_train, scoring=scoring, return_train_score=True)
scores_sss = cross_validate(mod2, X_train, y_train, scoring=scoring, return_train_score=True)
print('ROC_AUC')
print(scores_skf['train_ROC_AUC'])
print(scores_sss['train_ROC_AUC'])

print('F1-Weighted')
print(scores_skf['train_F1_weighted'])
print(scores_sss['train_F1_weighted'])
```

I will use <b>StratifiedShuffleSplit</b> as Corss validation startegy
