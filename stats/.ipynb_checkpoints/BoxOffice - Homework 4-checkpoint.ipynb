###  2.2 Plots

```r
#Original_language
ggplot(movies_ff, aes(movies_ff$original_language)) +
  geom_bar(color = "black") +  theme(text = element_text(size=120)) + theme_light()
options(repr.plot.width = 15, repr.plot.height = 10)
```

```r
#Runtime_cat
ggplot(movies_ff, aes(movies_ff$runtime_cat)) +
  geom_bar(fill = "#E69F00",color = "#E69F00",size = 0.3)  + theme_light()
options(repr.plot.width = 5, repr.plot.height = 3)
```

```r

par(mfrow = c(2, 2))
loop.vector <- 5:85
for (i in loop.vector){
    x <- movies_ff[,i]
    hist (x, xlab = i, main = colnames(movies_ff[,i]))
}
```

```r
hist(movies_ff$runtime,main = "runtime")
```

```r
# Budget
ggplot(movies_ff, aes(x=log(budget))) + 
 geom_histogram(aes(y=..density..), colour="black", fill="white")+
 geom_density(alpha=0.2, fill="#E69F00") + theme_get()
```

```r
#Popularity
ggplot(data=movies_ff) +
    geom_density(aes(popularity), main = "popularity")
```

```r
# Revenue
ggplot(movies_ff, aes(x=log(revenue))) + 
 geom_histogram(aes(y=..density..), colour="black", fill="white")+
 geom_density(alpha=0.2, fill="#E69F00") + theme_get()
```

```r
par(mfrow = c(2, 2))
loop.vector <- 9:12
for (i in loop.vector){
    x <- movies_ff[,i]
    hist(binomial (x))
    }
```
